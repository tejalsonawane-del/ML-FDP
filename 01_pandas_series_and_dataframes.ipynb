## Section 1: What Is Pandas and WHY Does It Matter for ML?

### The Problem: NumPy Is Not Enough

NumPy gives us fast numerical arrays, but it has two critical limitations for real-world ML:

1. **No labels** — NumPy arrays use only integer positions. Real data has column names like "age", "income", "diagnosis".
2. **Homogeneous types only** — A NumPy array holds ONE data type. Real datasets mix numbers, text, dates, and categories.

### The Solution: Pandas

Pandas gives us **labeled, heterogeneous** data structures built on top of NumPy:

| Feature | NumPy | Pandas |
|---------|-------|--------|
| Labels | No (integer index only) | Yes (named rows and columns) |
| Mixed types | No (one dtype per array) | Yes (different dtype per column) |
| Missing data | Limited (`np.nan`) | Built-in (`NaN`, `None`, `pd.NA`) |
| Data loading | Manual parsing | `read_csv`, `read_excel`, `read_sql` |
| Data exploration | Manual | `head()`, `info()`, `describe()` |

### The ML Pipeline

Every ML project follows this pipeline, and Pandas is at the center:

```
Raw Data (CSV, JSON, SQL)
    --> Pandas DataFrame (load, clean, explore, transform)
        --> X (features matrix) and y (target vector)
            --> ML Model (sklearn, PyTorch, TensorFlow)
```

**You will use Pandas in EVERY SINGLE ML project.** It is not optional — it is foundational.

### Let's import Pandas and NumPy

````python
# ============================================================
# Import pandas with the standard alias 'pd'
# Import numpy with the standard alias 'np'
# These two aliases are universal conventions — every tutorial,
# every codebase, every ML project uses them.
# ============================================================
import pandas as pd   # pd is the universal alias for pandas
import numpy as np    # np is the universal alias for numpy

# Print version info to verify installation
print("Pandas version:", pd.__version__)   # You need pandas >= 1.0
print("NumPy version: ", np.__version__)   # NumPy is pandas' foundation
print()
print("Both libraries imported successfully!")
print("We will use 'pd' instead of typing 'pandas' every time.")
print("We will use 'np' instead of typing 'numpy' every time.")
````

### The Bridge from NumPy to Pandas

In Module 00, you learned that:
- A **vector** is a 1D array — a single data sample (a point in feature space)
- A **matrix** is a 2D array — a collection of sample vectors (your entire dataset)

Pandas builds directly on top of this:

| NumPy Concept | Pandas Equivalent | What It Adds |
|---|---|---|
| 1D array (vector) | **Series** | Labels (index) + mixed types + NaN handling |
| 2D array (matrix) | **DataFrame** | Column names + row labels + per-column dtypes |

A DataFrame IS the labeled, heterogeneous version of the NumPy matrix you already know. Under the hood, each column of a DataFrame is stored as a NumPy array. When you call `df.to_numpy()`, you get back the raw NumPy matrix.

```
NumPy (Module 00):                    Pandas (Module 01):

np.array([[1500, 3, 350000],          pd.DataFrame({
          [2200, 4, 520000],              'sqft':  [1500, 2200, 800],
          [ 800, 1, 180000]])             'beds':  [3, 4, 1],
                                          'price': [350000, 520000, 180000]
   shape (3, 3)                       })            
   No column names                       Columns: sqft, beds, price
   All same dtype                        Each column can have its own dtype
   Integer index only                    Labeled rows and columns
```

**Think of it this way:** NumPy gives you the raw engine (fast array math). Pandas wraps that engine in a car with a dashboard, GPS, and seatbelts — the labels, exploration tools, and safety features you need for real-world data.

````python
# --- The NumPy-Pandas connection ---

# Create a NumPy matrix (what you learned in Module 00)
numpy_matrix = np.array([
    [1500,  3, 350000],
    [2200,  4, 520000],
    [ 800,  1, 180000]
])
print("NumPy matrix (no labels):")
print(numpy_matrix)
print(f"Shape: {numpy_matrix.shape}, dtype: {numpy_matrix.dtype}")
print()

# Wrap it in a DataFrame (add labels!)
pandas_df = pd.DataFrame(numpy_matrix, columns=['sqft', 'beds', 'price'])
print("Pandas DataFrame (labeled):")
print(pandas_df)
print()

# Go back to NumPy when needed (e.g., for sklearn)
back_to_numpy = pandas_df.to_numpy()
print("Back to NumPy with .to_numpy():")
print(back_to_numpy)
print(f"Same shape: {back_to_numpy.shape}")
print()
print("Each column IS a NumPy array underneath:")
print(f"  pandas_df['sqft'].values = {pandas_df['sqft'].values}")
print(f"  type: {type(pandas_df['sqft'].values)}")
````

## Section 2: Pandas Series

A **Series** is a one-dimensional array with **labels** (called an index). Think of it as a single column from a spreadsheet.

### WHY Series matter for ML:
- A single feature column in your dataset (e.g., "age") is a Series
- The target variable you are predicting (e.g., "price") is a Series
- Many pandas operations return a Series, so you must understand them

A Series has two parts:
1. **Values** — the actual data (stored as a NumPy array internally)
2. **Index** — labels for each value (default: 0, 1, 2, ...)

### 2.1 Creating a Series from a List

The simplest way to create a Series. Pandas automatically assigns integer indices starting from 0.

````python
# ============================================================
# Example 1: Clean demo — Series from a simple list
# Pandas assigns default integer index: 0, 1, 2, 3, 4
# ============================================================
temperatures = pd.Series([36.5, 37.0, 38.2, 36.8, 37.5])  # body temps in Celsius

print("=== Example 1: Series from a list (default index) ===")
print(temperatures)                    # Shows index on left, values on right
print()
print("Notice the LEFT column (0,1,2,3,4) — that is the INDEX.")
print("Notice the RIGHT column — those are the VALUES.")
print("dtype: float64 means these are 64-bit floating point numbers.")
````

````python
# ============================================================
# Example 2: Real-world ML context — Series with custom index
# In ML, your data often has meaningful labels (patient IDs,
# dates, sensor names). Custom indices let you access data
# by those labels instead of remembering positions.
# ============================================================
patient_temps = pd.Series(
    [36.5, 37.0, 38.2, 36.8, 37.5],                         # the data values
    index=['Patient_A', 'Patient_B', 'Patient_C',            # meaningful labels
           'Patient_D', 'Patient_E']
)

print("=== Example 2: Series with custom index (ML context) ===")
print(patient_temps)
print()
print("Now instead of remembering 'position 2', we can say 'Patient_C'.")
print("This makes our code self-documenting and less error-prone.")
print()
# Access by label — much more readable than by position
print("Patient_C's temperature:", patient_temps['Patient_C'])  # 38.2
print("Is Patient_C feverish?  ", patient_temps['Patient_C'] > 37.5)  # True
````

````python
# ============================================================
# Example 3: Edge case — Series with duplicate indices
# Yes, Pandas ALLOWS duplicate index labels!
# This can happen in real data and cause subtle bugs.
# ============================================================
duplicate_index = pd.Series(
    [100, 200, 300, 400],              # four values
    index=['A', 'B', 'A', 'B']        # duplicate labels: two A's, two B's
)

print("=== Example 3: Edge case — duplicate index labels ===")
print(duplicate_index)
print()
# Accessing 'A' returns MULTIPLE values (a Series, not a scalar)
print("duplicate_index['A'] returns:")
print(duplicate_index['A'])           # Returns a Series with 2 values
print()
print("Type of result:", type(duplicate_index['A']))  # <class 'pandas.core.series.Series'>
print()
print("WARNING: Duplicate indices are allowed but dangerous!")
print("When you expect a single value but get a Series, your code can break.")
print("Always check: series.index.is_unique =", duplicate_index.index.is_unique)
````

### 2.2 Creating a Series from a Dictionary

When you create a Series from a dict, the **keys become the index** and the **values become the data**. This is the most natural way to create labeled data.

````python
# ============================================================
# Example 1: Clean demo — dict keys become index labels
# ============================================================
model_accuracy = pd.Series({
    'Logistic Regression': 0.85,       # key -> index label
    'Random Forest': 0.92,             # value -> data value
    'SVM': 0.88,
    'Neural Network': 0.95
})

print("=== Series from a dictionary ===")
print(model_accuracy)
print()
print("Dictionary keys automatically became the index labels!")
print("This is perfect for storing named results like model scores.")
````

````python
# ============================================================
# Example 2: Real-world ML context — feature importances
# After training a model, you get feature importance scores.
# Storing them in a Series makes analysis easy.
# ============================================================
feature_importance = pd.Series({
    'age': 0.35,                       # age contributes 35% to prediction
    'income': 0.28,                    # income contributes 28%
    'education': 0.20,                 # education contributes 20%
    'location': 0.12,                  # location contributes 12%
    'gender': 0.05                     # gender contributes 5%
})
feature_importance.name = 'importance'  # Give the Series a name

print("=== Feature importance scores ===")
print(feature_importance)
print()
# Series supports sorting — instantly find top features
print("Top features (sorted):")
print(feature_importance.sort_values(ascending=False))  # Descending order
print()
print("Most important feature:", feature_importance.idxmax())  # 'age'
print("Least important feature:", feature_importance.idxmin())  # 'gender'
````

````python
# ============================================================
# Example 3: Edge case — dict with missing/None values
# Real data ALWAYS has missing values. Pandas handles them
# with NaN (Not a Number) or None.
# ============================================================
survey_responses = pd.Series({
    'Q1_satisfaction': 4,
    'Q2_recommend': None,              # respondent skipped this question
    'Q3_ease_of_use': 5,
    'Q4_value': np.nan,                # another missing value (NumPy style)
    'Q5_overall': 3
})

print("=== Series with missing values ===")
print(survey_responses)
print()
print("Notice: None and np.nan both show as NaN in the output.")
print("dtype is float64 (not int!) because NaN is a float value.")
print()
# Detecting missing values
print("Which values are missing?")
print(survey_responses.isna())         # True where value is NaN
print()
print("Number of missing values:", survey_responses.isna().sum())  # 2
print("Number of valid values:  ", survey_responses.notna().sum())  # 3
````

### 2.3 Series Attributes

Every Series has built-in attributes that tell you about its structure. These are properties, not methods — no parentheses needed.

**WHY these matter for ML:**
- `.dtype` tells you if your data is numeric (ready for ML) or text (needs encoding)
- `.shape` tells you how many samples you have
- `.index` lets you align data from different sources

````python
# ============================================================
# Create a Series to explore its attributes
# ============================================================
scores = pd.Series(
    [92, 85, 78, 95, 88],
    index=['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    name='exam_scores'                 # Series can have a name
)

print("=== Series Attributes ===")
print("The Series:")
print(scores)
print()

# .index — the labels (row names)
print("1. .index (the labels):")
print("   ", scores.index)                      # Index object
print("    As a list:", list(scores.index))      # Convert to Python list
print()

# .values — the data as a NumPy array
print("2. .values (the data as NumPy array):")
print("   ", scores.values)                      # array([92, 85, 78, 95, 88])
print("    Type:", type(scores.values))          # <class 'numpy.ndarray'>
print()

# .dtype — the data type of ALL values in the Series
print("3. .dtype (data type):")
print("   ", scores.dtype)                       # int64
print("    WHY this matters: ML models need numeric dtypes.")
print("    If dtype is 'object', you have text that needs encoding!")
print()

# .shape — dimensions as a tuple
print("4. .shape (dimensions):")
print("   ", scores.shape)                       # (5,)
print("    This means 5 elements in 1 dimension.")
print("    shape[0] =", scores.shape[0], "= number of samples")
print()

# .name — the name of the Series
print("5. .name:")
print("   ", scores.name)                        # 'exam_scores'
print("    When a Series becomes a DataFrame column, .name = column header")
print()

# .size — total number of elements (same as shape[0] for 1D)
print("6. .size:", scores.size)                  # 5
````

### 2.4 Accessing Elements: Label vs. Position

There are two ways to get data from a Series:
- **`series['label']`** or **`series.loc['label']`** — access by label name
- **`series.iloc[position]`** — access by integer position (0-based)

**WHY this distinction matters:** When your index is NOT the default 0,1,2,..., label-based and position-based access give DIFFERENT results.

````python
# ============================================================
# Example 1: Basic element access
# ============================================================
print("=== Accessing by LABEL (name) ===")
print("Alice's score:", scores['Alice'])          # 92 — access by label
print("Bob's score:  ", scores.loc['Bob'])        # 85 — explicit label access
print()

print("=== Accessing by POSITION (integer) ===")
print("First score (position 0):", scores.iloc[0])    # 92
print("Last score (position -1):", scores.iloc[-1])   # 88
print()

print("=== Accessing MULTIPLE elements ===")
# By labels — pass a list of labels
print("Alice and Diana (by label):")
print(scores[['Alice', 'Diana']])                 # Returns a sub-Series
print()

# By positions — pass a list of positions
print("First and third (by position):")
print(scores.iloc[[0, 2]])                        # Returns a sub-Series
````

````python
# ============================================================
# Example 2: Real-world ML context — slicing Series
# ============================================================
print("=== Slicing with .loc (INCLUSIVE of both endpoints) ===")
print("Bob through Diana:")
print(scores.loc['Bob':'Diana'])      # Includes both 'Bob' AND 'Diana'
print()

print("=== Slicing with .iloc (EXCLUSIVE of end) ===")
print("Positions 1 through 3 (excludes 3):")
print(scores.iloc[1:3])               # Includes position 1,2 but NOT 3
print()

print("KEY DIFFERENCE:")
print("  .loc['Bob':'Diana']  => includes Diana (INCLUSIVE end)")
print("  .iloc[1:3]           => excludes position 3 (EXCLUSIVE end)")
print("This is a common source of bugs! Remember which one you are using.")
````

````python
# ============================================================
# Example 3: Edge case — boolean indexing on Series
# This is HOW you filter data in ML preprocessing.
# ============================================================
print("=== Boolean indexing (filtering) ===")

# Create a boolean mask: which scores are >= 90?
high_scores_mask = scores >= 90       # Returns a Series of True/False
print("Mask (scores >= 90):")
print(high_scores_mask)
print()

# Use the mask to filter
high_scorers = scores[high_scores_mask]  # Only keeps True values
print("High scorers (>= 90):")
print(high_scorers)
print()
print("This is EXACTLY how you will filter ML datasets:")
print("  df[df['column'] > threshold]")
````

### 2.5 Vectorized Operations on Series

Just like NumPy, Pandas supports **vectorized operations** — applying math to ALL elements at once, without writing loops. This is both cleaner and MUCH faster than loops.

**WHY this matters for ML:** Feature engineering (creating new features from existing ones) relies heavily on vectorized operations.

````python
# ============================================================
# Example 1: Clean demo — arithmetic on entire Series
# ============================================================
print("=== Vectorized arithmetic ===")
print("Original scores:")
print(scores)
print()

# Add 5 bonus points to everyone
curved_scores = scores + 5             # Adds 5 to EVERY element
print("After adding 5 bonus points:")
print(curved_scores)
print()

# Scale scores to 0-1 range (min-max normalization)
# This is a REAL ML preprocessing step!
normalized = (scores - scores.min()) / (scores.max() - scores.min())
print("Normalized to 0-1 range (min-max scaling):")
print(normalized)
print()
print("WHY normalize? Many ML algorithms (KNN, SVM, Neural Nets)")
print("perform better when features are on the same scale.")
````

````python
# ============================================================
# Example 2: Real-world ML context — operations between Series
# ============================================================
math_scores = pd.Series({'Alice': 92, 'Bob': 85, 'Charlie': 78})
science_scores = pd.Series({'Alice': 88, 'Bob': 91, 'Charlie': 82})

# Element-wise operations align by INDEX, not by position
average = (math_scores + science_scores) / 2    # Aligns by name automatically
print("=== Operations between two Series ===")
print("Math scores:   ")
print(math_scores)
print()
print("Science scores:")
print(science_scores)
print()
print("Average scores (aligned by index):")
print(average)
print()
print("Pandas automatically matched Alice with Alice, Bob with Bob, etc.")
print("This INDEX ALIGNMENT is a superpower that NumPy does not have!")
````

````python
# ============================================================
# Example 3: Edge case — operations with mismatched indices
# When indices don't fully match, Pandas fills with NaN.
# ============================================================
set_a = pd.Series({'Alice': 92, 'Bob': 85, 'Charlie': 78})
set_b = pd.Series({'Alice': 88, 'Bob': 91, 'Diana': 95})  # Diana, not Charlie!

result = set_a + set_b                 # What happens to Charlie and Diana?
print("=== Mismatched indices produce NaN ===")
print("Set A:", dict(set_a))
print("Set B:", dict(set_b))
print()
print("set_a + set_b:")
print(result)
print()
print("Charlie is in A but not B => NaN")
print("Diana is in B but not A   => NaN")
print("Only Alice and Bob have values in BOTH => valid sum")
print()
print("WHY this matters: When merging data from different sources,")
print("mismatched indices cause silent NaN propagation. Always check!")
````

## Section 3: DataFrame Creation

A **DataFrame** is a 2D table with labeled rows and columns. It is the PRIMARY data structure you will use in ML.

Think of it as:
- A spreadsheet with named columns
- A SQL table
- A collection of Series that share the same index

### WHY DataFrames matter for ML:
- Your training data is a DataFrame: each **row** is a sample, each **column** is a feature
- You load CSVs into DataFrames, clean them, then convert to X (features) and y (target) for training
- Pandas DataFrames are the input to sklearn, XGBoost, and many other ML libraries

There are **3 main ways** to create a DataFrame. Each has its use case.

### 3.1 Method 1: Dictionary of Lists (Most Common)

Each key is a **column name**, each list is the **column data**. This is the most intuitive method and the one you will use most often.

**When to use:** When you are building a dataset column by column, or when your data is naturally organized by features.

````python
# ============================================================
# Example 1: Clean demo — basic dict of lists
# Each key = column name, each list = column values
# All lists MUST be the same length!
# ============================================================
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],   # 5 names
    'age': [25, 30, 35, 28, 32],                             # 5 ages
    'score': [92, 85, 78, 95, 88]                            # 5 scores
}

df_demo = pd.DataFrame(data)          # Create DataFrame from dict

print("=== DataFrame from dict of lists ===")
print(df_demo)
print()
print("Shape:", df_demo.shape)         # (5, 3) = 5 rows, 3 columns
print("Columns:", list(df_demo.columns))  # ['name', 'age', 'score']
print()
print("Each dictionary KEY became a COLUMN HEADER.")
print("Each dictionary VALUE (list) became a COLUMN of data.")
print("Pandas auto-assigned row index: 0, 1, 2, 3, 4.")
````

````python
# ============================================================
# Example 2: Real-world ML context — a training dataset
# This is EXACTLY how real ML datasets are structured:
#   - Each row = one sample/observation
#   - Each column = one feature or the target variable
# ============================================================
ml_dataset = pd.DataFrame({
    'house_size_sqft': [1400, 1600, 1700, 1875, 2100],    # Feature 1
    'num_bedrooms': [3, 3, 4, 4, 5],                       # Feature 2
    'age_years': [10, 15, 5, 20, 8],                       # Feature 3
    'has_garage': [1, 0, 1, 1, 1],                         # Feature 4 (binary)
    'price_usd': [250000, 280000, 310000, 295000, 380000]  # TARGET variable
})

print("=== ML Dataset: House Price Prediction ===")
print(ml_dataset)
print()
print("In ML terminology:")
print("  Rows    = samples (5 houses)")
print("  Columns = features + target")
print("  Features (X):", list(ml_dataset.columns[:-1]))    # All columns except last
print("  Target   (y):", ml_dataset.columns[-1])            # Last column
print()
# This is how you split features and target for ML:
X = ml_dataset.drop('price_usd', axis=1)  # All columns EXCEPT target
y = ml_dataset['price_usd']                # Only the target column
print("X (features) shape:", X.shape)       # (5, 4)
print("y (target) shape:  ", y.shape)       # (5,)
````

````python
# ============================================================
# Example 3: Edge case — mismatched list lengths cause errors
# ============================================================
print("=== Edge case: What if lists have different lengths? ===")
print()

try:
    bad_data = {
        'name': ['Alice', 'Bob', 'Charlie'],   # 3 elements
        'age': [25, 30,7]                         # 2 elements — MISMATCH!
    }
    bad_df = pd.DataFrame(bad_data)            # This will CRASH
except ValueError as e:
    print("ERROR:", e)                          # "All arrays must be of the same length"
    print()
    print("This error means your lists have different lengths.")
    print("Every column MUST have the same number of rows.")
    print("Fix: Make sure all lists have the same number of elements.")
````

### 3.2 Method 2: List of Lists (Row-Oriented)

Each inner list is a **row**. You must provide column names separately.

**When to use:** When your data is naturally organized by samples/records (e.g., reading data row by row from a file or database query).

````python
# ============================================================
# Example 1: Clean demo — list of lists with column names
# ============================================================
row_data = [
    ['Alice', 25, 92],                 # Row 0
    ['Bob', 30, 85],                   # Row 1
    ['Charlie', 35, 78],               # Row 2
    ['Diana', 28, 95],                 # Row 3
    ['Eve', 32, 88]                    # Row 4
]

# MUST specify column names — Pandas can't guess them from a list
df_from_lists = pd.DataFrame(row_data, columns=['name', 'age', 'score'])

print("=== DataFrame from list of lists ===")
print(df_from_lists)
print()
print("We HAD to provide column names with the 'columns' parameter.")
print("Without it, columns would be named 0, 1, 2 — not useful!")
````

````python
# ============================================================
# Example 2: Real-world ML context — sensor readings
# Imagine a sensor that logs [timestamp, temperature, humidity]
# each second. You collect readings as rows.
# ============================================================
sensor_readings = [
    ['2024-01-01 00:00', 22.5, 45.0],   # reading at midnight
    ['2024-01-01 00:01', 22.3, 45.2],   # reading at 00:01
    ['2024-01-01 00:02', 22.1, 45.5],   # reading at 00:02
    ['2024-01-01 00:03', 21.8, 46.0],   # temperature dropping
    ['2024-01-01 00:04', 21.5, 46.3],   # humidity rising
]

df_sensor = pd.DataFrame(
    sensor_readings,
    columns=['timestamp', 'temperature_c', 'humidity_pct']
)

print("=== Sensor data (row-oriented) ===")
print(df_sensor)
print()
print("Row-oriented creation is natural when data arrives one record at a time.")
print("This is common with: sensor data, log files, API responses, web scraping.")
````

````python
# ============================================================
# Example 3: Edge case — list of lists WITHOUT column names
# ============================================================
no_names = pd.DataFrame([
    [1, 2, 3],
    [4, 5, 6]
])

print("=== Without column names ===")
print(no_names)
print()
print("Columns are auto-named 0, 1, 2 — completely meaningless!")
print("ALWAYS provide column names when using list of lists.")
print()

# You can also set/rename columns after creation
no_names.columns = ['feature_1', 'feature_2', 'target']
print("After renaming columns:")
print(no_names)
````

### 3.3 Method 3: List of Dictionaries (JSON-Like)

Each dictionary is a **row**, and the keys become column names. This matches JSON format exactly.

**When to use:** When your data comes from JSON APIs, MongoDB queries, or any source that returns records as key-value pairs.

````python
# ============================================================
# Example 1: Clean demo — list of dicts
# Each dict is one row. Keys = column names.
# ============================================================
dict_data = [
    {'name': 'Alice', 'age': 25, 'score': 92},    # Row 0
    {'name': 'Bob', 'age': 30, 'score': 85},      # Row 1
    {'name': 'Charlie', 'age': 35, 'score': 78},  # Row 2
]

df_from_dicts = pd.DataFrame(dict_data)

print("=== DataFrame from list of dicts ===")
print(df_from_dicts)
print()
print("Dict keys automatically became column names — no 'columns' param needed!")
print("This is the format you get from JSON APIs and NoSQL databases.")
````

### 4.2 Loading the Iris Dataset from sklearn

The **Iris dataset** is the most famous dataset in machine learning. It contains measurements of 150 iris flowers from 3 species. We will use it throughout this course.

**What is it?**
- 150 samples (flowers)
- 4 features: sepal length, sepal width, petal length, petal width (all in cm)
- 3 target classes: setosa (0), versicolor (1), virginica (2)
- **Goal:** Predict the species from the measurements (classification task)

````python
# ============================================================
# Example 1: Load Iris and create a DataFrame
# ============================================================
from sklearn.datasets import load_iris   # Import the dataset loader

# Load the dataset — returns a Bunch object (like a dict)
iris = load_iris()

print("=== What does load_iris() return? ===")
print("Type:", type(iris))                 # sklearn.utils.Bunch
print("Keys:", list(iris.keys()))          # data, target, feature_names, etc.
print()

# Create a DataFrame from the features
df = pd.DataFrame(
    iris.data,                             # The feature values (150x4 NumPy array)
    columns=iris.feature_names             # The column names
)

# Add the target (species) as a new column
df['species'] = iris.target                # 0, 1, or 2

print("=== Iris DataFrame ===")
print(df.head(10))                         # Show first 10 rows
print()
print("Shape:", df.shape)                  # (150, 5)
print("  150 rows = 150 flower samples")
print("  5 columns = 4 features + 1 target")
````

````python
# ============================================================
# Example 2: Understanding the target variable
# ============================================================
# The target is encoded as integers: 0, 1, 2
# Let's add a human-readable species name column
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species_name'] = df['species'].map(species_map)  # Map integers to names

print("=== Target variable explained ===")
print("Target mapping:")
for code_val, name in species_map.items():
    print(f"  {code_val} = {name}")        # 0=setosa, 1=versicolor, 2=virginica
print()

print("First 3 rows with species name:")
print(df[['sepal length (cm)', 'species', 'species_name']].head(3))
print()

print("How many flowers per species?")
print(df['species_name'].value_counts())   # 50 each — perfectly balanced!
print()
print("A balanced dataset means equal samples per class.")
print("This is ideal for classification — no class imbalance issues.")
````

````python
# ============================================================
# Example 3: ML perspective — separating X and y
# This is a critical step you will do in EVERY ML project.
# ============================================================
# X = features matrix (what the model sees as input)
X = df[iris.feature_names]                 # Only the 4 measurement columns
# y = target vector (what the model tries to predict)
y = df['species']                          # The species integer code

print("=== Separating features (X) and target (y) ===")
print("X shape:", X.shape)                 # (150, 4) — 150 samples, 4 features
print("y shape:", y.shape)                 # (150,) — 150 labels
print()
print("X (first 5 rows):")
print(X.head())
print()
print("y (first 5 values):")
print(y.head())
print()
print("This X, y split is the STANDARD INPUT to sklearn:")
print("  model.fit(X, y)      # Train the model")
print("  predictions = model.predict(X)  # Make predictions")
````

### Quick Preview: Handling Missing Values

When `info()` reveals missing values, you will need to handle them before feeding data to an ML model — **most models crash on NaN**.

Here is a quick preview of the tools (covered in depth in Notebook 4):

```python
# Check: how many missing values per column?
df.isnull().sum()

# Option 1: Fill missing values (numeric → median, categorical → mode)
df['age'] = df['age'].fillna(df['age'].median())

# Option 2: Drop rows with missing values
df = df.dropna()
```

### 5.5 `describe()` — Statistical Summary

`describe()` computes summary statistics for all numeric columns: count, mean, std, min, quartiles, max.

**WHY each statistic matters for ML:**
- **count** — confirms how many non-null values (should match shape[0])
- **mean** — the average value; useful for imputation and feature understanding
- **std** — standard deviation; high std means high variance (may need scaling)
- **min/max** — the range; look for impossible values (age=999, price=-50)
- **25%/50%/75%** — quartiles; reveal the distribution shape and outliers

````python
# ============================================================
# Example 1: describe() on Iris dataset
# ============================================================
print("=== df.describe() — Statistical summary ===")
print(df.describe())
print()
print("How to read this for ML:")
print()
print("1. COUNT: All 150 — no missing values in numeric columns")
print("2. MEAN vs MEDIAN (50%):")
print("   sepal length mean=5.84, median=5.80 — close => roughly symmetric")
print("   When mean >> median, data is right-skewed (may need transformation)")
print()
print("3. STD: Petal measurements have higher std than sepal measurements.")
print("   Different scales => you may need FEATURE SCALING (StandardScaler)")
print()
print("4. MIN/MAX: Check for impossible values. All values here look reasonable.")
print("   sepal length: 4.3 to 7.9 cm — makes botanical sense")
print()
print("5. QUARTILES: 50% of sepal lengths fall between 5.1 (25%) and 6.4 (75%)")
print("   This is the Interquartile Range (IQR), used for outlier detection")
````

````python
# ============================================================
# Example 2: Real-world — describe() catches data problems
# ============================================================
problematic_data = pd.DataFrame({
    'age': [25, 30, 35, -5, 999, 28, 32, 40, 22, 27],  # -5 and 999 are errors
    'salary': [50000, 65000, 72000, 55000, 68000,
               0, 61000, 73000, 48000, 52000],            # 0 salary is suspicious
    'satisfaction': [4, 5, 3, 4, 5, 3, 2, 5, 4, 3]       # 1-5 scale, looks clean
})

print("=== describe() catches data quality issues ===")
print(problematic_data.describe())
print()
print("Red flags spotted:")
print("1. age min = -5 => impossible! Negative age is a data entry error")
print("2. age max = 999 => likely a placeholder for missing value")
print("3. salary min = 0 => suspicious, could be error or unpaid intern")
print()
print("describe() is your early warning system for data quality!")
````

### 5.6 `value_counts()` and `nunique()` — Understand Categorical Data

`value_counts()` counts occurrences of each unique value. `nunique()` counts how many unique values exist.

**WHY these matter for ML:**
- **Class balance:** If your target has 90% "No" and 10% "Yes", your model will just predict "No" always. You need to detect this!
- **Cardinality:** A column with 10,000 unique values needs different encoding than one with 3 unique values
- **Data quality:** Unexpected values reveal errors (e.g., "yse" instead of "yes")

````python
# ============================================================
# Example 1: value_counts() on the target variable
# ============================================================
print("=== value_counts() — Class distribution ===")
print(df['species_name'].value_counts())
print()
print("Each species has exactly 50 samples — this is PERFECTLY balanced!")
print("Balanced classes make classification easier and fairer.")
print()

# Also show as percentages (normalized)
print("As percentages:")
print(df['species_name'].value_counts(normalize=True))
print()
print("Each class is exactly 33.3% of the dataset.")
print("In real data, you rarely get this — imbalance is the norm.")
````

````python
# ============================================================
# Example 2: Real-world — detecting class imbalance
# ============================================================
# Simulate an imbalanced dataset (common in fraud detection, medical diagnosis)
imbalanced = pd.Series(
    ['normal'] * 950 + ['fraud'] * 50,   # 95% normal, 5% fraud
    name='transaction_type'
)

print("=== Imbalanced class distribution ===")
print(imbalanced.value_counts())
print()
print("Percentages:")
print(imbalanced.value_counts(normalize=True).round(4) * 100)
print()
print("WARNING: 95% normal, 5% fraud!")
print("A model that ALWAYS predicts 'normal' would be 95% accurate!")
print("But it would catch ZERO fraud cases — useless.")
print("Techniques to handle: oversampling, undersampling, class weights, SMOTE.")
````

````python
# ============================================================
# Example 3: nunique() — cardinality of each column
# ============================================================
print("=== nunique() — number of unique values per column ===")
print(df.nunique())
print()
print("Interpretation:")
print("  sepal length: 35 unique values out of 150 — continuous numeric")
print("  species:      3 unique values — categorical (the target)")
print("  species_name: 3 unique values — categorical (text label)")
print()
print("WHY cardinality matters:")
print("  Low cardinality (3-10 unique values) => one-hot encoding works well")
print("  High cardinality (100+ unique values) => one-hot creates too many columns")
print("    Use target encoding, embedding, or frequency encoding instead")
````
